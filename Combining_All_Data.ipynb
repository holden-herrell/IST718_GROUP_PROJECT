```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
```

```python
#Import Sports Data
sportURL="https://raw.githubusercontent.com/holden-herrell/IST718_GROUP_PROJECT/master/DATASETS/COMBINED%20DATA/Sports.csv"
Sports=pd.DataFrame(pd.read_csv(sportURL, engine='python', error_bad_lines=False))
```

```python
#Import Accident Data
atlURL="https://raw.githubusercontent.com/holden-herrell/IST718_GROUP_PROJECT/master/DATASETS/ACCIDENT%20DATA/ATL_Accidents.csv"
dalURL="https://raw.githubusercontent.com/holden-herrell/IST718_GROUP_PROJECT/master/DATASETS/ACCIDENT%20DATA/DAL_Accidents.csv"
houURL="https://raw.githubusercontent.com/holden-herrell/IST718_GROUP_PROJECT/master/DATASETS/ACCIDENT%20DATA/HOU_Accidents.csv"
miaURL="https://raw.githubusercontent.com/holden-herrell/IST718_GROUP_PROJECT/master/DATASETS/ACCIDENT%20DATA/MIA_Accidents.csv"
ATLraw=pd.read_csv(atlURL, engine='python', error_bad_lines=False)
ATL=pd.DataFrame(ATLraw)
DALraw=pd.read_csv(dalURL, engine='python', error_bad_lines=False)
DAL=pd.DataFrame(DALraw)
HOUraw=pd.read_csv(houURL, engine='python', error_bad_lines=False)
HOU=pd.DataFrame(HOUraw)
MIAraw=pd.read_csv(miaURL, engine='python', error_bad_lines=False)
MIA=pd.DataFrame(MIAraw)
```

```python
display(MIA)
```

```python
#Remove Unneeded Accident columns
#ATL=ATL.drop(columns=['ID','Description','County','Country','Airport_Code','TMC','Source','Street','Timezone'])
#DAL=DAL.drop(columns=['ID','Description','County','Country','Airport_Code','TMC','Source','Street','Timezone'])
#HOU=HOU.drop(columns=['ID','Description','County','Country','Airport_Code','TMC','Source','Street','Timezone'])
#MIA=MIA.drop(columns=['ID','Description','County','Country','Airport_Code','TMC','Source','Street','Timezone'])
```

```python
#Add Date Field
ATL['Date']=pd.to_datetime(ATL['Start_Time'], format= '%Y-%m-%d').dt.date
DAL['Date']=pd.to_datetime(DAL['Start_Time'], format= '%Y-%m-%d').dt.date
HOU['Date']=pd.to_datetime(HOU['Start_Time'], format= '%Y-%m-%d').dt.date
MIA['Date']=pd.to_datetime(MIA['Start_Time'], format= '%Y-%m-%d').dt.date
```

```python
#Calculate accident clean up time
ATL['CleanUpTime']=pd.to_datetime(ATL['End_Time'])-pd.to_datetime(ATL['Start_Time'])
DAL['CleanUpTime']=pd.to_datetime(DAL['End_Time'])-pd.to_datetime(DAL['Start_Time'])
HOU['CleanUpTime']=pd.to_datetime(HOU['End_Time'])-pd.to_datetime(HOU['Start_Time'])
MIA['CleanUpTime']=pd.to_datetime(MIA['End_Time'])-pd.to_datetime(MIA['Start_Time'])
```

```python
#Remove Extra Time columns
ATL=ATL.drop(columns=['Start_Time','End_Time','Weather_Timestamp'])
DAL=DAL.drop(columns=['Start_Time','End_Time','Weather_Timestamp'])
HOU=HOU.drop(columns=['Start_Time','End_Time','Weather_Timestamp'])
MIA=MIA.drop(columns=['Start_Time','End_Time','Weather_Timestamp'])
```

```python
display(MIA)
```

```python
#Join all accident data sets
ATL['JoinID']=ATL['Date'].astype(str)+ATL['City'].astype(str)+ATL['State'].astype(str)
DAL['JoinID']=DAL['Date'].astype(str)+DAL['City'].astype(str)+DAL['State'].astype(str)
HOU['JoinID']=HOU['Date'].astype(str)+HOU['City'].astype(str)+HOU['State'].astype(str)
MIA['JoinID']=MIA['Date'].astype(str)+MIA['City'].astype(str)+MIA['State'].astype(str)
US=pd.DataFrame(pd.concat([ATL,DAL,HOU,MIA], axis=0, join='outer', ignore_index=False, keys='JoinID',
          levels=None, names=None, verify_integrity=False, copy=True, sort=True))
```

```python
display(US)
```

```python
#Merge all datsets
USAgg=pd.merge(US,Sports, how='outer', on=['JoinID', 'JoinID'])
```

```python
display(USAgg)
```

```python
#Rename Columns, drop extras from join
USAgg.rename(columns={'City_x':'City','Date_x':'Date','State_x':'State'},inplace=True)
USAgg=USAgg.drop(columns=['City_y','Date_y','State_y', 'JoinID'])
```

```python
#Fill NA values for sports
USAgg['MLBGame'].fillna(0, inplace=True)
USAgg['NBAGame'].fillna(0, inplace=True)
USAgg['NFLGame'].fillna(0, inplace=True)
USAgg['MLB_Attendance'].fillna(0, inplace=True)
USAgg['NBA_Attendance'].fillna(0, inplace=True)
USAgg['NFL_Attendance'].fillna(0, inplace=True)
```

```python
display(USAgg)
```

```python
#Create consolidated lat, lng fields
USAgg['Lat']=USAgg['Start_Lat']
USAgg['Lng']=USAgg['Start_Lng']
USAgg.loc[USAgg['Lat']=='NaN', ['Lat']] = USAgg['End_Lat']
USAgg.loc[USAgg['Lng']=='NaN', ['Lng']] = USAgg['End_Lng']
USAgg=pd.DataFrame(USAgg)
USAgg=USAgg.drop(columns=['Start_Lat','Start_Lng','End_Lat','End_Lng'])
```

```python
pd.options.display.max_columns=60
pd.options.display.max_rows=200
display(USAgg)
```

```python
#Fill in null Day of Week
USAgg['Day of Week']=pd.to_datetime(USAgg['Date']).dt.day_name()
```

```python
#Remove records with too many NA values
USAgg=USAgg.dropna(how='any',subset=['City','State','Date','Lat','Lng'])
```

```python
display(USAgg)
```

```python
#Create Day/Night, remove redundant
USAgg=USAgg.rename(columns={'Sunrise_Sunset':'Day_or_Night','Day of Week':'Day_of_Week'})
#USAgg=USAgg.drop(columns=['Astronomical_Twilight','Civil_Twilight','Nautical_Twilight'])
```

```python
USAgg=USAgg[['Date','City', 'State', 'Zipcode', 'Lat', 'Lng','Day_of_Week',
       'MLBGame', 'MLB_Attendance', 'NBAGame', 'NBA_Attendance', 'NFLGame',
       'NFL_Attendance', 'CleanUpTime', 'Distance(mi)','Crossing','Amenity', 'Bump', 
        'Give_Way', 'Humidity(%)', 'Junction', 'No_Exit',
       'Number', 'Precipitation(in)', 'Pressure(in)', 'Railway', 'Roundabout',
       'Severity', 'Side', 'Station', 'Stop', 'Day_or_Night',
       'Temperature(F)', 'Traffic_Calming', 'Traffic_Signal', 'Turning_Loop',
       'Visibility(mi)', 'Weather_Condition', 'Wind_Chill(F)',
       'Wind_Direction', 'Wind_Speed(mph)' ]]
```
